#### Write a Function Problem
https://www.hackerrank.com/challenges/write-a-function/problem




```python
def is_leap(year):
    """The year can be evenly divided by 4, is a leap year, unless:
    The year can be evenly divided by 100, it is NOT a leap year, unless:
    The year is also evenly divisible by 400. Then it is a leap year."""
    # Write your logic here
    return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0) 
```

### Finding the percentage Problem
https://www.hackerrank.com/challenges/finding-the-percentage/problem



```python
def get_percetange(student, student_marks):
    marks = student_marks.get(student)
    if not marks: 
        return
    return sum(marks)/len(marks)
    

if __name__ == '__main__':
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()
    res = get_percetange(query_name, student_marks)
    print(f"{res:.2f}")
    
```

#### Python lists problem
https://www.hackerrank.com/challenges/python-lists/problem



```python
def build_commands(l):
    def append(val):
        l.append(val)
    def insert(i, val):
        l.insert(i, val)
    def printList():
        print(l)
    def pop():
        return l.pop()
    def reverse():
        return l.reverse()
    def sort():
        return l.sort()
    def remove(e):
        l.remove(e)
    return {
        "append": append,
        "insert": insert,
        "print": printList,
        "pop": pop,
        "reverse": reverse,
        "sort": sort,
        "remove": remove
    }

if __name__ == '__main__':
    N = int(input())
    l = []
    commands = build_commands(l)
    for i in range(N):
        command = input().split()
        params = len(command) > 1 and list(map(int, command[1:]))
        c = commands.get(command[0])
        if params: 
            c(*params)
        else: 
            c()
```

#### DefaultDict Tutorial
https://www.hackerrank.com/challenges/defaultdict-tutorial/problem



```python
# Enter your code here. Read input from STDIN. Print output to STDOUT

from collections import defaultdict

n, m = input().split(" ")

a = [input() for _ in range(int(n))]

b = [input() for _ in range(int(m))]

d = defaultdict(lambda: [])


for i, v in enumerate(a, start=1):
    d[v].append(i)

for v in b:
    if not d[v]:
        print(-1)
    else:
        print(*d[v]) 
```

#### Collections Counter Problem

https://www.hackerrank.com/challenges/collections-counter/problem



```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
from collections import Counter

l = input()
elements = input().split(" ")
inventory = Counter(elements)
earning = 0
for c in range(int(input())):
    ordered_qty, price = input().split(" ")
    qty = inventory[ordered_qty] 
    if(not qty or qty < 0):
        continue
    earning += int(price);
    inventory.subtract([ordered_qty])

print(earning)
```

#### Minimum Swaps 2
https://www.hackerrank.com/challenges/minimum-swaps-2/problem?h_l=interview&playlist_slugs%5B%5D=interview-preparation-kit&playlist_slugs%5B%5D=arrays



```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the minimumSwaps function below.
def minimumSwaps(arr):
    i = swaps = 0
    while i < len(arr):
        if arr[i] != i + 1:
            j = arr[i] - 1
            arr[i], arr[j] = arr[j], arr[i]
            swaps += 1
        if arr[i] == i + 1:
            i += 1
    return swaps
        
        
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input())

    arr = list(map(int, input().rstrip().split()))

    res = minimumSwaps(arr)

    fptr.write(str(res) + '\n')

    fptr.close()
```

#### Sorting Bubble Sort
https://www.hackerrank.com/challenges/ctci-bubble-sort/problem?h_l=interview&playlist_slugs%5B%5D=interview-preparation-kit&playlist_slugs%5B%5D=sorting



```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the countSwaps function below.
def countSwaps(a):
    i = j = swaps = 0
    while i < len(a):
        j = 0
        while j < len(a) - 1:
            if a[j] > a[j + 1]: 
                a[j], a[j + 1] = a[j + 1], a[j]
                swaps += 1
            j += 1
        i += 1
    print(f"Array is sorted in {swaps} swaps.")
    print(f"First Element: {a[0]}")
    print(f"Last Element: {a[-1]}")
    
if __name__ == '__main__':
    n = int(input())

    a = list(map(int, input().rstrip().split()))

    countSwaps(a)
```
