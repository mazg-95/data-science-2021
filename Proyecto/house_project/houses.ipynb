```python
def compute_y_hat(x, mb):
    return x.dot(mb)

def compute_error(y, y_hat):
    return ((y - y_hat) ** 2).mean() * 0.5
```

```python

def train_model(x_var, y_var, epochs=100, imprimir_error_cada=1, lr=0.01):
    mb = np.array([0,0], dtype=float)
    x = pd.DataFrame(data={ 'x': x_var, 't': np.ones(x_var.shape[0])})
    errors = [None] * epochs
    results = [mb] + [None] * epochs
    for epoch in range(epochs):
        y_hat = compute_y_hat(x, mb)
        errors[epoch] = compute_error(y_var, y_hat)
        if (epoch +1) % imprimir_error_cada == 0:
            print(f"{epoch + 1} Iteracion: (Error, {errors[epoch]}) ")
        grad = x.mul(y_hat - y_var, axis=0).mean()
        res = mb - grad.to_numpy()  * lr # Convert Series result to numpy array for consistent type. 
        results[epoch + 1] = mb  = res
    return pd.DataFrame(data=results, columns=['m', 'b']), errors  
```

```python
# Visualizacion del error en el tiempo.
def plot_errors(errors):
    fig, ax = plt.subplots()
    ax.plot(errors)
    ax.axis('auto')
    ax.set_title('Avg Error vs Iteration')
    ax.ticklabel_format(style='plain',axis='x',useOffset=False)
    
plot_errors([21061.125, 5542.57, 1792.18, 100, 50, 20, 30 ,40])
```

```python
# Visualizacion del modelo

def plot_trainning(model_history, x_var, y, n=1):
    model_params = model_history[(model_history.index + 1) % n == 0]
    x = np.transpose(np.column_stack((x_var.to_numpy(), np.ones(x_var.shape[0]))))
    y_hat_values = np.matmul(model_params, x)
    fig, ax = plt.subplots(model_params.shape[0],1, figsize=(8, 6 * model_params.shape[0]))
    for index, y_hat in y_hat_values.iterrows():
        i = ((index+1) // n) - 1
        x_vals = x[0,:]
        ax[i].plot(x_vals, y_hat, color='r')
        ax[i].scatter(x_vals, y)
        ax[i].set_title(f"Iteration :{index + 1}")
    fig.tight_layout()
    plt.show()
```

```python
### Modelo Overall Quality
x_var, y_var = train_df["OvrllQual"], train_df["SalePrice"]
model_history_1, errors = train_model(x_var, y_var, epochs=10000, imprimir_error_cada=500, lr=0.0499)

plot_errors(errors)

plot_trainning(model_history_1, x_var, y_var, n=500)
```

```python
def predict(manual_model, sk_model, x_test):
    x = np.column_stack((x_test.to_numpy(), np.ones(x_test.shape[0])))
    mb = manual_model.to_numpy()
    res_manual = compute_y_hat(x, mb)
    res_sk = sk_model.predict(x[:,0].reshape(-1,1))
    res_avg = (res_manual + res_sk) * 0.5
    return res_manual, res_sk, res_avg

test_df = pd.DataFrame(test_data, columns=["SalePrice", "OvrllQual", "1stFlrSF", "TotRmAbvGrd", "YearBuilt" ,"LotFrontage"])
```
